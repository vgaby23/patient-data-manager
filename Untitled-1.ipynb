```python
query5 = """
SELECT status AS status_distribution,
       COUNT(*) AS total_count
FROM appointments
GROUP BY status
"""
df_5 = run_sql(query5)
```

```python
plt.figure()
plt.pie(df_5['total_count'], 
        labels = df_5['status_distribution'],
        autopct = '%1.1f%%',
        startangle = 90)
plt.title('Appointment Status Distribution')
```

```python
query6 = """
SELECT DATE_FORMAT(appointment_date, '%m') AS month,
       COUNT(*) AS total_appointments,
       SUM(CASE WHEN LOWER(TRIM(status)) = 'no-show' THEN 1 ELSE 0 END) AS no_show_count,
       ROUND(SUM(CASE WHEN LOWER(TRIM(status)) = 'no-show' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS percentage_no_show
FROM appointments
GROUP BY month
ORDER BY month ASC
"""
no_show = run_sql(query6)
```

```python
# Calculate successful appointments for stacking
attended = [t - n for t, n in zip(no_show['total_appointments'], no_show['no_show_count'])]

fig, ax1 = plt.subplots(figsize=(10, 6))

# Primary Axis: Stacked Bars for volume
ax1.bar(no_show['month'], attended, label='Attended', color='#4f81bd')
ax1.bar(no_show['month'], no_show['no_show_count'], bottom=attended, label='No-Show', color='#c0504d')
ax1.set_ylabel('Number of Appointments', color='black')
ax1.set_xlabel('Month')
ax1.legend(loc='upper left')
ax1.grid(axis='y', linestyle='--', alpha=0.5)

# Secondary Axis: Line for Percentage Rate
ax2 = ax1.twinx()
ax2.plot(no_show['month'], no_show['percentage_no_show'], color='#e36c09', marker='o', linewidth=2.5, label='No-Show %')
ax2.set_ylabel('No-Show Rate (%)', color='#e36c09')
ax2.tick_params(axis='y', labelcolor='#e36c09')
ax2.legend(loc='upper right')

plt.title('Monthly Appointment Volumes vs. No-Show Rates')
plt.show()
```

```python
query7 = """
SELECT 
    payment_status,
    COUNT(bill_id) AS total_bills
FROM bills
GROUP BY payment_status
HAVING COUNT(bill_id) > 10
ORDER BY total_bills DESC;
"""
df_7 = run_sql(query7)
```

```python
query15 = """
SELECT 
    d.doctor_id,
    d.first_name,
    d.last_name,
    d.specialization,
    DATE_FORMAT(treatment_date, '%m') AS month,
    SUM(t.cost) AS total_treatment_cost,
    COUNT(t.treatment_id) AS number_of_treatments
FROM doctors d
JOIN appointments a ON d.doctor_id = a.doctor_id
JOIN treatments t ON a.appointment_id = t.appointment_id
JOIN bills b on b.treatment_id = t.treatment_id
WHERE b.payment_status = 'Paid'
GROUP BY d.doctor_id, month
ORDER BY month ASC;
"""
df_15 = run_sql(query15)
```

```python
df_15['doctor_name'] = df_15['first_name'] + ' ' + df_15['last_name']

# Pivot the data so rows=doctors, columns=months, values=total_cost
heatmap_data = df_15.pivot(index='doctor_name', columns='month', values='total_treatment_cost')

plt.figure(figsize=(12, 8))
# cmap='YlGnBu' creates a clear light-to-dark color progression
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)

plt.title('Doctor Revenue Heatmap across Months ($)')
plt.xlabel('Month')
plt.ylabel('Doctor')
plt.tight_layout()
plt.show()
```

```python
# 2. Group by doctor and sum their total revenue over the year
top_doctors = (df_15.groupby('doctor_name')['total_treatment_cost']
               .sum()
               .reset_index()
               .sort_values(by='total_treatment_cost', ascending=False))

# 3. Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_doctors, 
    x='total_treatment_cost', 
    y='doctor_name', 
    palette='Blues_r'  # Darker blue for the top performer
)

plt.title('Top Performing Doctors Over the Year (By Total Revenue)')
plt.xlabel('Total Revenue Generated ($)')
plt.ylabel('Doctor Name')
plt.tight_layout()
plt.show()
```

```python
query3 = """
with
first_visit as 
    (select patient_id, min(appointment_date) as first_visit
    from appointments 
    group by patient_id)
select *, date_format(appointment_date, '%m') as month_appointment,
    case when appointment_date = first_visit then 'New Patient' else 'Returning Patient'
    end as patient_type
from appointments a
join first_visit fv on a.patient_id = fv.patient_id
"""
df_3 = run_sql(query3)

monthly_retention = (df_3.groupby(['month_appointment', 'patient_type'])
                     .size()
                     .unstack(fill_value=0))

# Convert periods to strings for cleaner plotting on the X-axis
monthly_retention.index = monthly_retention.index.astype(str)

# 6. Plot the Stacked Area Chart
colors = ['#2ca02c', '#1f77b4'] # Green for New, Blue for Returning
monthly_retention.plot(kind='area', stacked=True, figsize=(12, 6), color=colors, alpha=0.8)

plt.title('Hospital Growth: New vs. Returning Patients Month-over-Month')
plt.xlabel('Month')
plt.ylabel('Total Number of Appointments')
plt.legend(title='Patient Type', loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
